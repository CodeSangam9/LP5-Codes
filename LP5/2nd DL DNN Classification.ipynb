```python
# =========================================================
# IMDB Binary Classification (No Matplotlib Version)
# =========================================================

import numpy as np
import warnings
warnings.filterwarnings("ignore")

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.datasets import imdb
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.metrics import classification_report

print("TensorFlow version:", tf.__version__)

# Parameters
NUM_WORDS = 10000
MAX_LEN = 256

# ----------------------------
# Load dataset  ✅ FIXED
# ----------------------------
(x_train_raw, y_train), (x_test_raw, y_test) = imdb.load_data(num_words=NUM_WORDS)

# ----------------------------
# Multi-hot encoding
# ----------------------------
def multi_hot(sequences):
    result = np.zeros((len(sequences), NUM_WORDS))
    for i, seq in enumerate(sequences):
        result[i, seq] = 1
    return result

x_train = multi_hot(x_train_raw)
x_test = multi_hot(x_test_raw)

# ----------------------------
# Split validation
# ----------------------------
val_size = int(0.2 * len(x_train))

x_val, x_train = x_train[:val_size], x_train[val_size:]
y_val, y_train = y_train[:val_size], y_train[val_size:]

# ----------------------------
# Model
# ----------------------------
model = keras.Sequential([
    layers.Input(shape=(NUM_WORDS,)),
    layers.Dense(64, activation='relu'),
    layers.Dropout(0.4),
    layers.Dense(64, activation='relu'),
    layers.Dense(1, activation='sigmoid')
])

model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

# ----------------------------
# Train
# ----------------------------
history = model.fit(
    x_train, y_train,
    epochs=10,
    batch_size=512,
    validation_data=(x_val, y_val),
    callbacks=[EarlyStopping(patience=2)]
)

# ----------------------------
# Evaluate
# ----------------------------
loss, acc = model.evaluate(x_test, y_test)

print("\nTest Accuracy:", acc)

# ----------------------------
# Predictions
# ----------------------------
y_pred = (model.predict(x_test) > 0.5).astype(int)

print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```
